```python
import pandas as pd
winefile = pd.read_csv('C:/Users\Windows/Documents/machine_learning_winedataset/wine_dataset.csv')
```

```python
winefile
```

```python
winefile['style'] = winefile['style'].replace('red', 0)
```

```python
winefile['style'] = winefile['style'].replace('white', 1)
```

```python
winefile
```

```python
#separete variable between predictores and target
y = winefile['style']
x = winefile.drop('style', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

#create data for train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
#Model criation;
model = ExtraTreesClassifier(n_estimators=100)
model.fit(x_train,y_train)

#print results
result = model.score(x_test, y_test)
print("Accuracy:", result)
```
